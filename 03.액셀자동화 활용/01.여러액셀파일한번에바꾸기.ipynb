```python
""" 
[실습과제]
여러개의 액셀 파일을 바꾸는 프로그램을 만들어 본다.  
세금계산서_놀부전자.xlsx
[C:\Users\프로젝트폴더\공유폴더\계산서폴더]
위 디렉토리 안 5개의 액셀파일에 대해 '자료입력페이지' 시트의 '작성일자'를 '2030-02-01'로 한번에 수정되도록 프로그램을 제작해보자.
 """
```

```python
# 한개의 액셀파일 자동화
import xlwings as xw, os, glob, shutil
app = xw.App(add_book=False)
wb = app.books.open(r'C:\Users\프로젝트폴더\공유폴더\계산서파일\세금계산서_놀부전자.xlsx')
ws = wb.sheets['자료입력페이지']
ws.range('C14').value = '2030-02-01'
wb.save()
app.quit()
```

```python
# 파일 목록 추출
for file in glob.glob(r'C:\Users\프로젝트폴더\공유폴더\계산서파일\*.xlsx'):
  print(file)
```

```python
# 추출된 파일 목록에 액셀 자동화
for file in glob.glob(r'C:\Users\프로젝트폴더\공유폴더\계산서파일\*.xlsx'):
  app = xw.App(add_book=False)
  wb = app.books.open(file)
  ws = wb.sheets['자료입력페이지']
  ws.range('C14').value = '2030-02-01'
  wb.save()
  app.quit()
```

```python
# 
""" 
[다중 파일 액셀 자동화 최적화]
액셀 앱을 만들고 다시 닫는 과정을 반복문 안에서 반복하지 않고  
앱 만들기, 닫기는 for문 바깥에 작성한다
 """
app = xw.App(add_book=False)
for file in glob.glob(r'C:\Users\프로젝트폴더\공유폴더\계산서파일\*.xlsx'):
  wb = app.books.open(file)
  ws = wb.sheets['자료입력페이지']
  ws.range('C14').value = '2030-02-01'
  wb.save()
app.quit()
```
