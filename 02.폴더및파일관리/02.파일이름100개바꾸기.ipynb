```python
# 파일 100개 생성

import random
for i in range(1, 101):
  f = open(f'참이슬/판매_데이터{i}.txt', 'w', encoding = 'utf-8')
  f.write(f'{random.randint(1, 100)}병')
  f.close()
```

```python
# 파일 100개 이름 수정하기
# os.rename('수정전 경로', '수정후 경로')
# 판매_데이터1.txt -> sale_data1.txt
import os
for i in range(1, 101):
  os.rename(f'참이슬/판매_데이터{i}.txt', f'참이슬/sale_data{i}.txt')
```
